```python
# Read the dataset
dataset = pd.read_csv(r"/content/drive/MyDrive/CSC791:NLP/project/dataset.csv")
combined = pd.read_csv(r"/content/drive/MyDrive/CSC791:NLP/project/combined-features.csv")

# Separate the X(jokes) and the Y(is_humor or not)

jokes = list(dataset['text'])
labels = list(dataset['humor'])

# jokes = jokes[:1000]
# labels = labels[:1000]

# combined = combined.head(1000)
# dataset = dataset.head(1000)
```

```python
df = dataset
data = dataset
```

```python
train_texts, test_texts, train_labels, test_labels = train_test_split(data['text'], data['humor'], test_size=0.2, random_state=42)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Tokenize the text data
train_encodings = tokenizer(train_texts.tolist(), truncation=True, padding=True)
test_encodings = tokenizer(test_texts.tolist(), truncation=True, padding=True)
```

```python
# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    train_labels
)).shuffle(len(train_texts)).batch(8)

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    test_labels
)).batch(8)
```

```python
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
```

```python
# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=3e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metric = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')
```

```python
model.compile(optimizer=optimizer, loss=loss, metrics=[metric])
```

```python
model.fit(train_dataset, epochs=1, validation_data=test_dataset)
```

```python
model.evaluate(test_dataset)
```
